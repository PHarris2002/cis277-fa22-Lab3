## Cleaning of data:

```python
#drop duplicates
df.drop_duplicates(inplace=True)

#fix types/values
df.rating.replace("*9.0", 9.0, inplace=True)
df.rating = pd.to_numeric(df.rating)

df['votes'] = df['votes'].str.replace(',','')
df.votes = pd.to_numeric(df.votes)

#rename columns

df.columns = ['game_index', 'name', 'url', 'year', 'certificate', 'rating', 'votes',
       'plot_summary', 'Action', 'Adventure', 'Comedy', 'Crime', 'Family',
       'Fantasy', 'Mystery', 'Sci_Fi', 'Thriller']
```

## Questions

### 1. What is the average rating for Action games?

```python
action_average_rating = df[df.Action].rating.mean()

print(f"The average rating for action movies is {action_average_rating}.")
```

### 2. What is the average rating of games after 2010? What is the average rating of games before (and including) 2010?

```python
after_year_mask = df[df.year > 2010]
after_year_rating_average = after_year_mask.rating.mean()

print(f"The average rating for years after 2010 is {after_year_rating_average}")

before_year_mask = df[df.year <= 2010]
before_year_rating_average = before_year_mask.rating.mean()

print(f"The average rating for years before and during 2010 is {before_year_rating_average}")
```

### 3. What games include "Fallout" in their title?

```python
fallout_mask = df['name'].str.contains('Fallout')
fallout_movie_list = df[fallout_mask]

print("The movies below are all Fallout games from the game list.")
display(fallout_movie_list)
```

### 4. What is the most common certificate for Sci-Fi games? 

```python
certificate_count = sci_fi_max = df[df.Sci_Fi].certificate.value_counts()

display(certificate_count)
```

### 5. What percent of Thriller games have M or AO certificate?

```python
#Gets total of the entire sum
thriller_total = df['Thriller'].sum()

#Gets only Thriller's certificates
thriller_certificate_mask = df[df.Thriller].certificate

#Gets total of M and AO rated games
thriller_m_ao = df[(df.Thriller == True) & (df.certificate=='M') | (df.certificate=='AO')]
thriller_m_ao_total = len(thriller_m_ao)

#Calculates the percentage of M and AO rated games
calculated_percentage = (thriller_m_ao_total / thriller_total) * 100

print(f"{calculated_percentage}% of Thriller games have a M or AO certificate.")
```

```python
thriller_m_ao = df[(df.Thriller == True) & (df.certificate=='M') | (df.certificate=='AO')]

thriller = df[(df.Thriller == True)]

percent = (len(thriller_m_ao) / len(thriller)) * 100

print(f"{percent}% of Thriller games have a M or AO certificate.")
```

### 8. What is the average rating for each genre (Action, Adventure, Comedy, Crime, Family, Fantasy, Mystery, Sci-Fi, and Thriller)?

## 12. What is the most common certificate of games with ratings greater than 8.0 and votes greater than 1000?

```python
rating_and_vote_mask = (df.votes > 1000) & (df.rating > 8.0)

df[rating_and_vote_mask].certificate.value_counts()
```

## 13. For games with ratings greater than the 75th percentile for ratings and a number of votes greater than the 75th percentile for votes, what is the most common genre?

```python
rating_threshold = df.rating.quantile(0.75)
votes_threshold = df.votes.quantile(0.75)

good_games = df[(df.rating > rating_threshold) & (df.votes > votes_threshold)]

genres = ["Action", "Adventure", "Comedy", "Crime", "Family", "Fantasy", "Mystery", "Sci_Fi", "Thriller"]

for genre in genres:
    print(genre, len(good_games[good_games[genre]==True]))
```
